```r
# load libraries
library(tidyverse)
library(lubridate)
library(tidycensus)
```

```r
# bring in the data
ca_fha <- read_csv("ca_fha.csv")
```

```r
# look at data
# note that KPCC cleaned the down_payment_src vector and created the date vector prior to sharing
glimpse(ca_fha)
```

```r
# compute and graph the monthly trend
ca_fha %>% 
    group_by(month=floor_date(date, "month"), down_payment_src) %>% 
    tally() %>%
    ggplot(aes(x=month, y=n, fill=down_payment_src)) + geom_area(alpha=0.6 , size=1, colour="black")
```

```r
# get dataframe with monthly percentage of family-assisted loans, create plot of treand
monthly_src <- ca_fha %>% group_by(month=floor_date(date, "month"), down_payment_src) %>% tally()
monthly_src <- monthly_src %>% group_by(month) %>% summarise(monthly_total = sum(n)) %>% right_join(monthly_src, by = "month")
monthly_src$pct <- monthly_src$n / monthly_src$monthly_total
monthly_src %>% filter(down_payment_src == "Relative") %>% ggplot(aes(x=month, y=pct, group=1)) + geom_line(lwd=2, alpha=0.9)
```
